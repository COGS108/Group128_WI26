# Research Question

Is there a statistically significant difference in the prevalence of clinical depression (PHQ-9 scores) and anxiety (GAD-7 scores) among UC San Diego undergraduates compared to the UC System-wide average and National US University benchmarks (as defined by the Healthy Minds Study 2024-2025)? Furthermore, to what extent does 'Sense of Belonging' (measured via the 6-point UCUES Likert scale) serve as a predictor for these mental health outcomes when controlling for academic year and financial stress.



## Background and Prior Work

Students' mental health has become a central concern in the field of higher education, because psychological distress affects not only students' quality of life, but also their academic performance and their behaviors to seek help. To monitor indications such as depression symptoms, anxiety, stress, sense of belonging, and usage of counseling  service, campuses use standardized survey tools. However, different campuses measure mental health and well-being in different ways. As a result, we must carefully design metric selection and consider survey design variation, including sampling strategies, response rates and timing within a academic year, these are necessary for the studies.

This project will focus on asking how UCSD students' mental well-being and overall happiness compare to other UC campuses, other US universities, and univeristy internationlly. It can be framed as descriptive analysis(UCSD vs other UC campuses) and explanatory modeling(what student characteristics or campus conditions predict differences, after adjusting for confounders like student demographics, financial stress, or academic workload).
There are prior researches and work show that student mental health issues are common and institutional variables can affect the prevalence of symtoms and the likelyhood of seeking help.

Healthy Minds Study(HMS)<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

One of the most widely used systems for comparing mental health status in US colleges and universities is the Healthy Minds Study (HMS).The study surveyed students at participating institutions and reported prevalence rates of mental health indicators, help-seeking behaviors, and demand for services.Results from HMS indicate that mental health challenges are common among US college students. More importantly, the study shows a consistent gap between the number of students experiencing mental health difficulties and the number who actually seek treatment. This suggests that differences between campuses may reflect variations in access to care, stigma, or awareness, rather than differences in mental health alone.

However, prior work also notes that self-reported mental health measures may be influenced by response bias, campus culture, and differences in survey participation rates. As a result, higher reported prevalence at some campuses may reflect greater awareness or willingness to disclose symptoms rather than worse underlying mental health.

ACHA-NCHA<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

This is one of the largest and most comprehensive surveys of college student health in the United States. Conducted over multiple years across numerous institutions, the reference group summaries released by ACHA enable colleges to compare their student data against national benchmarks. These summaries encompass assessment metrics related to mental health, well-being, sense of belonging, and health behaviors.

Research results indicate that well-being is influenced both by mental health status and by broader campus experiences, such as a sense of belonging and perceptions of institutional support. This supports a perspective that views well-being as multidimensional rather than solely focused on mental health symptoms.

# Hypothesis


We hypothesize that there is a statistically significant difference in student mental well-being between UC San Diego and other UC campuses as well as U.S. universities overall, as measured by standardized indicators such as self-reported stress levels and happiness. Specifically, we predict that UC San Diego students will report comparable or slightly higher levels of psychological distress relative to peer institutions, reflecting differences in academic workload, campus culture, and perceived support systems.

This hypothesis is motivated by prior findings from large-scale surveys such as the Healthy Minds Study and ACHA-NCHA, which suggest that institutional characteristics and campus environments can meaningfully influence student mental health outcomes. We further expect that observed differences may reflect not only underlying mental health status, but also variation in awareness, willingness to disclose symptoms, and access to mental health resources across campuses.


## Data

### Data overview

For this project, we are not collecting our own survey data. Instead, we are using established campus and national survey sources that already measure student well-being using standardized questions. This matters for our research question because it lets us compare UC San Diego to other UC campuses and to broader U.S. benchmarks without inventing a new measurement scale.

**1) Ideal dataset**

In an ideal world, we would have a single dataset that includes multiple universities (UCSD, other UC campuses, and a wide range of U.S. schools) and measures mental well-being with the exact same survey items, administered during the same time window each year.

**Variables we would want:**
- Campus identifier (UCSD vs other UC campuses vs non-UC U.S. universities)
- Well-being outcomes: self-reported stress, anxiety/depression indicators, self-reported happiness or life satisfaction, and sense of belonging
- Help-seeking / resource use: whether a student used counseling/mental health services and perceived access/barriers
- Timing variables: quarter/semester timing and survey administration window (so we can check whether “finals season” effects are showing up)
- Basic controls: year in school, age range, gender (if available), major/college (if available), and financial stress indicators

**How many observations:**
To make cross-campus comparisons stable, we would want at least a few hundred responses per campus, and ideally 1,000+ per campus when possible. If sample sizes vary by campus, we would prefer weighted estimates or to focus comparisons on campuses with enough responses to avoid noisy results.

**How the data would be collected:**
The cleanest design would be a standardized, anonymous online survey administered to representative samples across campuses at the same point in the academic year. Ideally, campuses would use the same validated scales so that differences reflect real student experiences rather than differences in survey wording.

**How the data would be stored/organized:**
We would store the data in a tidy, tabular format where each row is one respondent and each column is one variable. If we only have campus-level summaries, the dataset would instead have one row per campus per year with summary statistics.

**2) Real datasets we will use**

**Healthy Minds Study (HMS), 2024–2025 National Data Report**
HMS provides national data on student mental health and help-seeking patterns. We will use it to benchmark UCSD and the UC system against broader U.S. college patterns. Key variables include stress/distress measures and help-seeking behavior.

**ACHA–NCHA (Spring 2025 reference summaries)**
ACHA–NCHA provides standardized college health survey metrics across many institutions. We will use it for national and reference group comparisons. Important variables include mental health indicators and well-being measures.

**UC Undergraduate Experience Survey (UCUES)**
UCUES is designed specifically for cross-campus comparison within the UC system. This will be our primary dataset for comparing UCSD to other UC campuses on self-reported well-being and stress. We prioritize UCUES because it reduces measurement inconsistency across campuses.

### Healthy Minds Study (HMS) – Student Mental Health Survey Data

1. This data set is from the Healthy Minds Study (HMS), which is a large-scale survey adminstarted across US colleges and universities to assess the mental health of students. The data set includes student survey responses, including standardized mental health screening instruments, demographic characteristics, and data on use of counseling services. The survey is usually conducted online during a term.
2. Important metrics
        The primary mental health outcomes in this dataset include:

        Depression symptoms (PHQ-9 score)
        Unit: Score from 0–27
        Interpretation:

        0–4: Non - Minimal depression

        5–9: Mild depression

        10–14: Moderate depression

        15–19: Moderately severe

        20–27: Severe depression

        Anxiety symptoms (GAD-7 score)
        Unit: Score from 0–21
        Interpretation:

        0–4: Non-Minimal anxiety

        5–9: Mild anxiety

        10–14: Moderate anxiety

        15–21: Moderate - Severe anxiety

        Life satisfaction / wellbeing score
        Often measured on a scale of 1–5 or 0–10.
        Higher values indicate greater subjective wellbeing.
        
        Counseling service use
        Binary (0 = No, 1 = Yes) or categorical (Never, Past year, Past month).
3. Major Concerns:
    Self-report bias
    Students tend to underreport or overreport their symptoms based on stigma or personal awareness.

    Self-selection bias
    Voluntary participation. Students who choose to participate may systematically differ from those who do not participate (e.g., more distressed students tend to participate).

    Non-response bias
    Some groups (e.g., international students, students with heavy workload) tend to have lower response rates.

    Timing effects
    Surveys taken around the time of midterms or finals may artificially inflate levels of stress.

```python
import pandas as pd
import numpy as np

path = "/home/l2deng/Project/Group128_WI26/data/02-processed/HMS_2024-25_cleaned.csv"
df = pd.read_csv(path)

print("Loaded:", path)
display(df.head())

print("\nSIZE")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nTIDY CHECK")
print("Tidy definition: 1 row = 1 student observation; 1 column = 1 variable.")
if "student_id" in df.columns:
    print("student_id unique:", df["student_id"].is_unique)
else:
    print("student_id column not found; cannot verify 1 row per student_id.")

print("Duplicate rows:", df.duplicated().sum())

print("\nMISSINGNESS")
missing_counts = df.isna().sum().sort_values(ascending=False)
missing_percent = (df.isna().mean() * 100).round(2).sort_values(ascending=False)

print("Columns with missing values:")
display(pd.DataFrame({"missing_count": missing_counts, "missing_percent": missing_percent}).query("missing_count > 0"))

missing_cols = missing_counts[missing_counts > 0].index.tolist()

if missing_cols:
    key_cols = [c for c in ["campus", "gender", "academic_year", "financial_stress"] if c in df.columns]

    if key_cols:
        for col in missing_cols[:10]:
            print(f"\nMissingness pattern for {col} (mean missing rate by group):")
            miss_ind = df[col].isna().astype(int)
            for g in key_cols:
                rates = miss_ind.groupby(df[g]).mean().sort_values(ascending=False)
                print(f"\n  by {g}:")
                print(rates)

    print("\nMissingness interpretation:")
    print(
        "Missingness appears low overall in this processed dataset. "
        "If missing rates are similar across campus/gender/etc., that supports MCAR/weak MAR. "
        "If a column's missing rate changes a lot by subgroup, that suggests systematic missingness (MAR)."
    )
else:
    print("No missing values detected.")

print("\nOUTLIERS / SUSPICIOUS ENTRIES")

def flag_out_of_range(series, lo, hi):
    return (series < lo) | (series > hi)

flags = pd.DataFrame(index=df.index)

if "PHQ9" in df.columns:
    flags["PHQ9_out_of_range"] = flag_out_of_range(df["PHQ9"], 0, 27)
if "GAD7" in df.columns:
    flags["GAD7_out_of_range"] = flag_out_of_range(df["GAD7"], 0, 21)
if "flourishing_score" in df.columns:
    flags["flourishing_out_of_range"] = flag_out_of_range(df["flourishing_score"], 8, 56)

def iqr_outliers(series):
    s = series.dropna()
    if s.empty:
        return pd.Series(False, index=series.index)
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    lo = q1 - 1.5 * iqr
    hi = q3 + 1.5 * iqr
    return (series < lo) | (series > hi)

if "PHQ9" in df.columns:
    flags["PHQ9_iqr_outlier"] = iqr_outliers(df["PHQ9"])
if "GAD7" in df.columns:
    flags["GAD7_iqr_outlier"] = iqr_outliers(df["GAD7"])
if "flourishing_score" in df.columns:
    flags["flourishing_iqr_outlier"] = iqr_outliers(df["flourishing_score"])

if {"PHQ9", "GAD7"}.issubset(df.columns):
    flags["dep_or_anx"] = ((df["PHQ9"] >= 10) | (df["GAD7"] >= 10))
if {"suicide_attempt_past_year", "suicidal_ideation_past_year"}.issubset(df.columns):
    flags["attempt_without_ideation"] = (df["suicide_attempt_past_year"] == 1) & (df["suicidal_ideation_past_year"] == 0)
if {"PHQ9", "suicidal_ideation_past_year"}.issubset(df.columns):
    flags["severe_phq9_no_ideation"] = (df["PHQ9"] >= 20) & (df["suicidal_ideation_past_year"] == 0)
if {"dep_or_anx", "therapy_past_year"}.issubset(df.columns):
    flags["no_symptoms_but_therapy"] = (~flags["dep_or_anx"]) & (df["therapy_past_year"] == 1)

flag_cols = [c for c in flags.columns if c != "dep_or_anx"]
flags["any_flag"] = flags[flag_cols].any(axis=1) if flag_cols else False

print("Flag counts:")
display(flags[flag_cols + ["any_flag"]].sum().sort_values(ascending=False).to_frame("count"))

if flags["any_flag"].any():
    print("\nExample flagged rows (first 10):")
    display(df.loc[flags["any_flag"], :].head(10))

print("\nCLEANING")

df_clean = df.copy()

core_outcomes = [c for c in ["PHQ9", "GAD7"] if c in df_clean.columns]
if core_outcomes:
    before = df_clean.shape[0]
    df_clean = df_clean.dropna(subset=core_outcomes, how="any")
    after = df_clean.shape[0]
    print(f"Dropped rows missing core outcomes {core_outcomes} (how='any'): {before - after} rows removed.")
    print(
        "Justification: PHQ-9 and GAD-7 are primary outcomes; rows missing them can't be used in prevalence/regression. "
        "Given low missingness, listwise deletion should have minimal impact."
    )
else:
    print("Core outcomes not found; no outcome-based dropping applied.")

range_cols = []
if "PHQ9" in df_clean.columns:
    range_cols.append(("PHQ9", 0, 27))
if "GAD7" in df_clean.columns:
    range_cols.append(("GAD7", 0, 21))
if "flourishing_score" in df_clean.columns:
    range_cols.append(("flourishing_score", 8, 56))

for col, lo, hi in range_cols:
    invalid = (df_clean[col] < lo) | (df_clean[col] > hi)
    if invalid.any():
        n = int(invalid.sum())
        df_clean.loc[invalid, col] = np.nan
        print(f"Set {n} invalid {col} values to NaN (outside [{lo}, {hi}]).")

print("\nFinal cleaned dataset size:", df_clean.shape)
print("Remaining missing values (top 15):")
display(df_clean.isna().sum().sort_values(ascending=False).head(15))
```

```python
import pandas as pd
import numpy as np

path = "/home/l2deng/Project/Group128_WI26/data/02-processed/NCHA_2025.csv"
df = pd.read_csv(path)

print("Loaded:", path)
display(df.head())

print("\nSIZE")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nTIDY CHECK")
print("Tidy definition: 1 row = 1 respondent observation; 1 column = 1 variable.")
if "respondent_id" in df.columns:
    print("respondent_id unique:", df["respondent_id"].is_unique)
else:
    print("respondent_id column not found; cannot verify 1 row per respondent_id.")

print("Duplicate rows:", df.duplicated().sum())

print("\nMISSINGNESS")
missing_counts = df.isna().sum().sort_values(ascending=False)
missing_percent = (df.isna().mean() * 100).round(2).sort_values(ascending=False)

print("Columns with missing values:")
display(pd.DataFrame({"missing_count": missing_counts, "missing_percent": missing_percent}).query("missing_count > 0"))

missing_cols = missing_counts[missing_counts > 0].index.tolist()

if missing_cols:
    key_cols = [c for c in ["sex_group_RSEX", "self_rated_health", "food_security_level", "stress_level_30d"] if c in df.columns]

    if key_cols:
        for col in missing_cols[:10]:
            print(f"\nMissingness pattern for {col} (mean missing rate by group):")
            miss_ind = df[col].isna().astype(int)
            for g in key_cols:
                rates = miss_ind.groupby(df[g]).mean().sort_values(ascending=False)
                print(f"\n  by {g}:")
                print(rates)

    print("\nMissingness interpretation:")
    print(
        "Missingness appears low overall. "
        "If missing rates are similar across sex/health/food security/stress groups, that supports MCAR/weak MAR. "
        "If a column's missing rate changes a lot by subgroup, that suggests systematic missingness (MAR)."
    )
else:
    print("No missing values detected.")

print("\nOUTLIERS / SUSPICIOUS ENTRIES")

def flag_out_of_range(series, lo, hi):
    return (series < lo) | (series > hi)

flags = pd.DataFrame(index=df.index)

if "k6_distress_score" in df.columns:
    flags["k6_out_of_range"] = flag_out_of_range(df["k6_distress_score"], 0, 24)
if "uls3_loneliness_score" in df.columns:
    flags["uls3_out_of_range"] = flag_out_of_range(df["uls3_loneliness_score"], 3, 9)
if "pwb_score" in df.columns:
    flags["pwb_out_of_range"] = flag_out_of_range(df["pwb_score"], 8, 56)
if "cdrisc2_resilience_score" in df.columns:
    flags["cdrisc2_out_of_range"] = flag_out_of_range(df["cdrisc2_resilience_score"], 0, 8)
if "sbqr_screening_score" in df.columns:
    flags["sbqr_out_of_range"] = flag_out_of_range(df["sbqr_screening_score"], 3, 18)

def iqr_outliers(series):
    s = series.dropna()
    if s.empty:
        return pd.Series(False, index=series.index)
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    lo = q1 - 1.5 * iqr
    hi = q3 + 1.5 * iqr
    return (series < lo) | (series > hi)

for col in ["k6_distress_score", "uls3_loneliness_score", "pwb_score", "cdrisc2_resilience_score", "sbqr_screening_score"]:
    if col in df.columns:
        flags[f"{col}_iqr_outlier"] = iqr_outliers(df[col])

if {"k6_distress_score", "serious_psychological_distress"}.issubset(df.columns):
    flags["serious_distress_inconsistent"] = ((df["k6_distress_score"] >= 13) & (df["serious_psychological_distress"] == 0)) | ((df["k6_distress_score"] < 13) & (df["serious_psychological_distress"] == 1))

if {"uls3_loneliness_score", "loneliness_positive"}.issubset(df.columns):
    flags["loneliness_inconsistent"] = ((df["uls3_loneliness_score"] >= 6) & (df["loneliness_positive"] == 0)) | ((df["uls3_loneliness_score"] < 6) & (df["loneliness_positive"] == 1))

flag_cols = [c for c in flags.columns]
flags["any_flag"] = flags[flag_cols].any(axis=1) if flag_cols else False

print("Flag counts:")
display(flags[flag_cols + ["any_flag"]].sum().sort_values(ascending=False).to_frame("count"))

if flags["any_flag"].any():
    print("\nExample flagged rows (first 10):")
    display(df.loc[flags["any_flag"], :].head(10))

print("\nCLEANING")

df_clean = df.copy()

core_outcomes = [c for c in ["k6_distress_score", "uls3_loneliness_score"] if c in df_clean.columns]
if core_outcomes:
    before = df_clean.shape[0]
    df_clean = df_clean.dropna(subset=core_outcomes, how="any")
    after = df_clean.shape[0]
    print(f"Dropped rows missing core outcomes {core_outcomes} (how='any'): {before - after} rows removed.")
    print(
        "Justification: K6 distress and ULS3 loneliness are primary outcomes for wellbeing. "
        "Rows missing them cannot be used in prevalence/regression; missingness is low so listwise deletion is reasonable."
    )
else:
    print("Core outcomes not found; no outcome-based dropping applied.")

range_cols = []
if "k6_distress_score" in df_clean.columns:
    range_cols.append(("k6_distress_score", 0, 24))
if "uls3_loneliness_score" in df_clean.columns:
    range_cols.append(("uls3_loneliness_score", 3, 9))
if "pwb_score" in df_clean.columns:
    range_cols.append(("pwb_score", 8, 56))
if "cdrisc2_resilience_score" in df_clean.columns:
    range_cols.append(("cdrisc2_resilience_score", 0, 8))
if "sbqr_screening_score" in df_clean.columns:
    range_cols.append(("sbqr_screening_score", 3, 18))

for col, lo, hi in range_cols:
    invalid = (df_clean[col] < lo) | (df_clean[col] > hi)
    if invalid.any():
        n_bad = int(invalid.sum())
        df_clean.loc[invalid, col] = np.nan
        print(f"Set {n_bad} invalid {col} values to NaN (outside [{lo}, {hi}]).")

if "food_security_level" in df_clean.columns:
    df_clean["food_security_level"] = df_clean["food_security_level"].fillna("Unknown")

print("\nFinal cleaned dataset size:", df_clean.shape)
print("Remaining missing values (top 15):")
display(df_clean.isna().sum().sort_values(ascending=False).head(15))
```

# UC Dataset #3
1. This data set is a summarized table, not a list of individual students, and it is an aggregation of a data set containing a survey of University of California students on their health, designated as an NCHAIII file. The data provides a count and percentage of responses on a variety of mental health questions related to stress (nervousness, effects of anxiety/depression/stress) and University of California Academic Experiences.
2. The key metrics/fields used in this CSV are:

   Count of NCHAIII: number of responses in each category (frequency).

   % of Total Count of NCHAIII along Mental List: percent distribution for the “mental list” items (e.g., nervousness frequency).

   1 2 3 4 5 + Grand Total: distribution across a 5-level academic-performance/impact scale (shown as percentages in “Calculation1…” rows and as counts in “Total…” rows).

   Calculation1 along Pivot Field Values / Total along Pivot Field Values (in Unnamed: 1): tells  whether that row is showing percentages or raw counts for the 1–5 scale.
3. Primary mental-health-related outcomes (by source_table) include:

   Felt Nervous in Past 30 Days (frequency categories like none → all of the time)
   
   Impact of Anxiety on Students (whether anxiety affected academics / performance)
   
   Impact of Depression on Students (whether depression affected academics / performance)
   
   Impact of Stress on Students (whether stress affected academics / performance)
   
   (“UC Academic Experiences” is more about academic challenges; it’s useful context, but it isn’t a mental-health screener outcome.)

```python
import pandas as pd
import numpy as np
import os

path = "/home/l2deng/Project/Group128_WI26/data/01-interim/UC_Merged.csv"
df_raw = pd.read_csv(path)

print("Loaded:", path)
display(df_raw.head())

print("\nSIZE (raw)")
print("Rows:", df_raw.shape[0])
print("Columns:", df_raw.shape[1])

def norm_str(x):
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    s = " ".join(s.split())
    return s if s != "" else np.nan

def to_number(x):
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    if s == "":
        return np.nan
    s = s.replace(",", "")
    if s.endswith("%"):
        try:
            return float(s[:-1]) / 100.0
        except:
            return np.nan
    try:
        return float(s)
    except:
        return np.nan

df = df_raw.copy()
df.columns = [norm_str(c) for c in df.columns]

for c in df.columns:
    if df[c].dtype == "object":
        df[c] = df[c].map(norm_str)

print("\nTIDY CHECK (context-specific)")
print("This UC-merged file is not student-level. It is a set of summary tables that contains data from all UC campuses.")
print("we make it tidy by converting to long format where:")
print("1 row = 1 measurement: (table/question, response category or scale option, value type) -> value")

opt_cols = [c for c in ["1", "2", "3", "4", "5", "Grand Total"] if c in df.columns]
has_acad_block = (
    ("Unnamed: 1" in df.columns)
    and ("Unnamed: 0" in df.columns)
    and (len(opt_cols) > 0)
    and df["Unnamed: 1"].notna().any()
)

tidy_frames = []

if has_acad_block:
    acad = df[df["Unnamed: 1"].notna()].copy()
    acad_long = acad.melt(
        id_vars=["source_table", "Unnamed: 0", "Unnamed: 1"],
        value_vars=opt_cols,
        var_name="response_option",
        value_name="raw_value",
    )
    acad_long["question"] = acad_long["Unnamed: 0"].map(norm_str)
    acad_long["row_label"] = acad_long["Unnamed: 1"].map(norm_str)
    acad_long["value_type"] = np.where(
        acad_long["row_label"].str.lower().str.contains("calculation", na=False),
        "percent",
        np.where(
            acad_long["row_label"].str.lower().str.contains("total", na=False),
            "count",
            "value",
        ),
    )
    acad_long["value"] = acad_long["raw_value"].map(to_number)
    acad_long["response"] = np.nan
    acad_long["scale"] = "1-5"
    tidy_frames.append(
        acad_long[
            ["source_table", "question", "response", "response_option", "scale", "value_type", "value"]
        ]
    )

dist = df[df["Unnamed: 1"].isna()].copy()
dist["response"] = dist["Unnamed: 0"].map(norm_str) if "Unnamed: 0" in df.columns else np.nan

percent_cols = [c for c in df.columns if ("%" in c) or ("of total" in c.lower()) or ("percent" in c.lower())]
count_cols = [c for c in df.columns if ("count" in c.lower()) and (c not in percent_cols)]

def add_dist_block(value_col, value_type):
    if value_col in dist.columns:
        tmp = dist[["source_table", "response"]].copy()
        tmp["question"] = tmp["source_table"].map(norm_str)
        tmp["response_option"] = np.nan
        tmp["scale"] = np.nan
        tmp["value_type"] = value_type
        tmp["value"] = dist[value_col].map(to_number)
        tidy_frames.append(
            tmp[["source_table", "question", "response", "response_option", "scale", "value_type", "value"]]
        )

for c in percent_cols:
    add_dist_block(c, "percent")
for c in count_cols:
    add_dist_block(c, "count")

tidy = pd.concat(tidy_frames, ignore_index=True)

print("\nSIZE (tidy)")
print("Rows:", tidy.shape[0])
print("Columns:", tidy.shape[1])
display(tidy.head(15))

print("\nMISSINGNESS (tidy)")
missing_counts = tidy.isna().sum().sort_values(ascending=False)
missing_percent = (tidy.isna().mean() * 100).round(2).sort_values(ascending=False)
display(pd.DataFrame({"missing_count": missing_counts, "missing_percent": missing_percent}).query("missing_count > 0"))

print("\nMissingness pattern (value missing rate) by table and value_type:")
miss_ind = tidy["value"].isna().astype(int)
display(
    miss_ind.groupby([tidy["source_table"], tidy["value_type"]])
    .mean()
    .sort_values(ascending=False)
    .to_frame("missing_rate")
)

print("\nOUTLIERS / SUSPICIOUS ENTRIES (tidy)")
flags = pd.DataFrame(index=tidy.index)

flags["percent_out_of_range"] = tidy["value_type"].eq("percent") & ((tidy["value"] < 0) | (tidy["value"] > 1))
flags["count_negative"] = tidy["value_type"].eq("count") & (tidy["value"] < 0)

ro = pd.to_numeric(tidy["response_option"], errors="coerce")
flags["response_option_out_of_range"] = ro.notna() & ((ro < 1) | (ro > 5)) & (~tidy["response_option"].eq("Grand Total"))

flags["any_flag"] = flags.any(axis=1)

print("Flag counts:")
display(flags.sum().sort_values(ascending=False).to_frame("count"))

if flags["any_flag"].any():
    print("\nExample flagged rows (first 15):")
    display(tidy.loc[flags["any_flag"], :].head(15))

print("\nCLEANING (tidy)")
tidy_clean = tidy.copy()

tidy_clean = tidy_clean.dropna(subset=["source_table", "question", "value_type"], how="any")

is_percent = tidy_clean["value_type"].eq("percent")
v = tidy_clean.loc[is_percent, "value"]
fix_mask = v.notna() & (v > 1) & (v <= 100)
tidy_clean.loc[is_percent & fix_mask, "value"] = tidy_clean.loc[is_percent & fix_mask, "value"] / 100.0

tidy_clean.loc[tidy_clean["value_type"].eq("percent") & ((tidy_clean["value"] < 0) | (tidy_clean["value"] > 1)), "value"] = np.nan
tidy_clean.loc[tidy_clean["value_type"].eq("count") & (tidy_clean["value"] < 0), "value"] = np.nan

before = tidy_clean.shape[0]
tidy_clean = tidy_clean.dropna(subset=["value"], how="all")
after = tidy_clean.shape[0]

tidy_clean = tidy_clean.drop_duplicates()

print("Dropped rows with missing value after cleaning:", before - after)
print("Justification: This UC dataset is aggregated summary output; rows without a numeric value are not analyzable,")
print("and invalid percents/counts are set to NaN to avoid contaminating comparisons.")

print("\nFinal cleaned tidy dataset size:", tidy_clean.shape)
print("Remaining missing values (top 10):")
display(tidy_clean.isna().sum().sort_values(ascending=False).head(10))
Summary = """The UC_Merged_tidy_clean.csv file is tidy and clean for what it represents: an aggregated UC summary export, not student-level data. After reshaping, it is tidy because each row is one measurement (a specific table/question × a response level or Likert option × a metric), and each variable is stored in its own column (source_table, question, response or response_option, value_type, value). The large “missingness” in response vs. response_option/scale is mostly structural—some tables use text categories (so response is filled), while the academic-experiences table uses a 1–5 scale (so response_option/scale are filled instead). Cleaning made the numeric fields usable by converting percent strings (e.g., 26%) into decimals (0.26), converting comma-separated counts (e.g., 5,523) into numbers, removing duplicates, and ensuring there are no invalid values (percents outside 0–1 or negative counts). Interpreting the data: it reports distributions of experiences and impacts (e.g., how often students felt nervous, or how anxiety/depression/stress impacted academics) and Likert-scale response distributions for academic stressors, provided as both counts and percentages."""
print("\nSummary")
print(Summary)

out_path = os.path.join(os.path.dirname(path), "UC_Merged_tidy_clean.csv")
if os.path.dirname(out_path):
    os.makedirs(os.path.dirname(out_path), exist_ok=True)

tidy_clean.to_csv(out_path, index=False)
print("\nWrote:", out_path)
```

#### Exploratory Data Analysis

The purpose of this exploratory data analysis (EDA) is to investigate patterns in student mental health indicators before conducting formal statistical testing. Our research question examines whether there are meaningful differences in mental well-being between UC San Diego students and students from other universities. In particular, we focus on indicators of psychological distress, loneliness, and related well-being measures.

Two datasets are used in this analysis. The first dataset contains individual-level responses from the National College Health Assessment survey, including standardized mental health indicators. The second dataset contains aggregated summary tables from UC campuses describing distributions of student experiences and mental health-related outcomes. Examining both datasets allows us to understand mental health patterns at both the individual student level and the institutional level.

We first examine the individual-level survey dataset containing student mental health indicators. The initial dataset contains 5,000 respondents and 25 variables, where each row represents a single student observation. A tidy data check confirms that the dataset follows a standard structure with one row per respondent and one column per variable. The unique respondent identifier confirms that there are no duplicate observations.

[Missing Data]
We evaluated missingness across the key mental health indicators used in the analysis. Missingness levels were relatively low across variables:

- K6 psychological distress score: 150 missing values (3.0%)
- ULS-3 loneliness score: 100 missing values (2.0%)
- Psychological well-being score: 75 missing values (1.5%)
- Resilience score: 50 missing values (1.0%)

To determine whether missingness was systematic, we compared missing rates across several subgroup variables including gender identity, self-rated health, food security status, and recent stress levels. Missingness rates were relatively similar across these groups, suggesting that the missing data are likely Missing Completely at Random (MCAR) or only weakly related to observable characteristics.

Because psychological distress and loneliness are the primary outcomes of interest, rows missing either variable were removed from the dataset using listwise deletion. This resulted in 247 rows being removed, leaving a final cleaned dataset of 4,753 observations and 25 variables.


```python
import pandas as pd
import numpy as np

path = "data/02-processed/NCHA_2025.csv"
df = pd.read_csv(path)

print("Dataset shape:", df.shape)
print("Unique respondents:", df["respondent_id"].nunique())
print("Duplicate rows:", df.duplicated().sum())

# Key mental health indicators
key_vars = [
    "k6_distress_score",
    "uls3_loneliness_score",
    "pwb_score",
    "cdrisc2_resilience_score"
]

# Ensure numeric
for c in key_vars:
    df[c] = pd.to_numeric(df[c], errors="coerce")

# Missingness table
missing_table = pd.DataFrame({
    "missing_count": df[key_vars].isna().sum(),
    "missing_percent": (df[key_vars].isna().mean()*100).round(2)
})

print("\nMissingness summary:")
display(missing_table)

# Check missingness across subgroups
group_vars = [
    "sex_group_RSEX",
    "self_rated_health",
    "food_security_level",
    "stress_level_30d"
]

print("\nMissingness by subgroup (K6 distress):")
for g in group_vars:
    if g in df.columns:
        rates = df["k6_distress_score"].isna().groupby(df[g]).mean()
        print("\n", g)
        display(rates)

# Remove rows missing primary outcomes
before = df.shape[0]

df_clean = df.dropna(subset=[
    "k6_distress_score",
    "uls3_loneliness_score"
])

after = df_clean.shape[0]

print("\nRows removed due to missing K6 or ULS3:", before-after)
print("Clean dataset shape:", df_clean.shape)
```

#### Section 3 of EDA - Cross-Campus Comparison (Bivariate Analysis)
To directly address our research question, we segmented the Healthy Minds Study (HMS) dataset into three categories: **UCSD**, **Other UCs** (e.g., UCLA, UC Berkeley), and **Other US Universities**. 

The boxplots below display the distribution of clinical depression (PHQ-9) and anxiety (GAD-7) scores across these three groups. This bivariate analysis allows us to visually inspect if UC San Diego students report higher median psychological distress compared to the national and UC-wide benchmarks before we move into formal hypothesis testing.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the HMS data since it contains the specific PHQ9 and GAD7 scores across campuses
hms_path = "/home/l2deng/Project/Group128_WI26/data/02-processed/HMS_2024-25_cleaned.csv"
df_hms = pd.read_csv(hms_path)

# Drop rows missing the core outcomes for this specific analysis
df_hms = df_hms.dropna(subset=['PHQ9', 'GAD7'])

# Create a categorization function to group the campuses for our research question
def categorize_campus(campus_name):
    if campus_name == 'UCSD':
        return 'UCSD'
    elif 'UC ' in str(campus_name) or campus_name == 'UCLA':
        return 'Other UCs'
    else:
        return 'Other US Universities'

# Apply the categorization
df_hms['Campus_Category'] = df_hms['campus'].apply(categorize_campus)

# Set up the matplotlib figure
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 1. Boxplot for Depression (PHQ-9)
sns.boxplot(data=df_hms, x='Campus_Category', y='PHQ9', ax=axes[0], palette='Set2')
axes[0].set_title('Distribution of Depression (PHQ-9) Scores')
axes[0].set_ylabel('PHQ-9 Score (Higher = More Depressed)')
axes[0].set_xlabel('Institution Type')

# 2. Boxplot for Anxiety (GAD-7)
sns.boxplot(data=df_hms, x='Campus_Category', y='GAD7', ax=axes[1], palette='Set3')
axes[1].set_title('Distribution of Anxiety (GAD-7) Scores')
axes[1].set_ylabel('GAD-7 Score (Higher = More Anxious)')
axes[1].set_xlabel('Institution Type')

plt.tight_layout()
plt.show()

# Display summary statistics to back up the plots
print("Mean Mental Health Scores by Campus Group:")
display(df_hms.groupby('Campus_Category')[['PHQ9', 'GAD7']].mean().round(2))
```
