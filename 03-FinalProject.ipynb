## Abstract

This project investigates whether UC San Diego students differ in levels of depression and anxiety compared to other UC campuses and national benchmarks, and whether a student's sense of belonging predicts mental health outcomes. Using large-scale survey datasets including the Healthy Minds Study (HMS), ACHA–NCHA, and UCUES, we analyzed standardized measures such as PHQ-9, GAD-7, and belonging scores.

We first ensured data quality and comparability across datasets, then conducted descriptive and inferential analyses including prevalence estimation and regression modeling. Our models examined the relationship between belonging and mental health outcomes while controlling for academic year and financial stress.

The results show that depression and anxiety levels at UCSD are generally comparable to peer institutions, with slight variation depending on the benchmark used. More importantly, sense of belonging emerges as a strong and consistent predictor of both depression and anxiety, even after accounting for key confounders. Students with lower belonging scores tend to report significantly higher psychological distress.

Overall, these findings suggest that student mental health is not solely an individual clinical issue but is closely tied to campus environment and social integration. Improving students' sense of belonging may be an effective pathway for institutions to support mental well-being at scale.

To monitor indications such as depression symptoms, anxiety, stress, sense of belonging, and usage of counseling service, campuses use standardized survey tools. However, different campuses measure mental health and well-being in different ways. This project will focus on asking how UCSD students' mental well-being and overall happiness compare to other UC campuses, other US universities, and univeristy internationlly. ### Clinical Prevalence and Group Mean Estimates for Depression and Anxiety

Our inferential-style result focuses on the prevalence of elevated depression and anxiety symptoms in the HMS sample. Because PHQ-9 and GAD-7 are validated screening tools with widely used thresholds, we can describe both average symptom burden and the proportion of students whose scores reach the clinically meaningful cutoff of 10 or above. This gives a more interpretable picture than reporting only raw means.


```python
if 'analysis_df' in locals():
    phq_clinical = analysis_df[phq_col] >= 10
    gad_clinical = analysis_df[gad_col] >= 10

    overall_prevalence = pd.DataFrame({
        'metric': ['Mean PHQ-9', 'Mean GAD-7', 'PHQ-9 >= 10', 'GAD-7 >= 10'],
        'value': [
            analysis_df[phq_col].mean(),
            analysis_df[gad_col].mean(),
            phq_clinical.mean(),
            gad_clinical.mean()
        ]
    })

    print('Overall depression and anxiety estimates:')
    display(overall_prevalence)

    if school_col is not None:
        school_summary = analysis_df.groupby(school_col).agg(
            n=(phq_col, 'size'),
            mean_phq=(phq_col, 'mean'),
            mean_gad=(gad_col, 'mean')
        )
        school_summary['phq_ge_10'] = analysis_df.groupby(school_col)[phq_col].apply(lambda s: (s >= 10).mean())
        school_summary['gad_ge_10'] = analysis_df.groupby(school_col)[gad_col].apply(lambda s: (s >= 10).mean())
        school_summary = school_summary.sort_values('n', ascending=False)
        print('\nSchool-level summary (first 10 rows):')
        display(school_summary.head(10))

    fig, axes = plt.subplots(1, 2, figsize=(10, 4))

    overall_rates = pd.Series({
        'PHQ-9 >= 10': phq_clinical.mean(),
        'GAD-7 >= 10': gad_clinical.mean()
    })
    overall_rates.plot(kind='bar', ax=axes[0])
    axes[0].set_title('Clinical-Range Prevalence')
    axes[0].set_ylabel('Proportion of students')
    axes[0].set_ylim(0, max(0.1, overall_rates.max() * 1.2))

    axes[1].boxplot([
        analysis_df[phq_col].dropna(),
        analysis_df[gad_col].dropna()
    ], labels=['PHQ-9', 'GAD-7'])
    axes[1].set_title('Symptom Score Spread')
    axes[1].set_ylabel('Score')

    plt.tight_layout()
    plt.show()
```

### Belonging as a Predictor of Depression and Anxiety, Controlling for Academic Year and Financial Stress

The central explanatory question of this project is whether a student's sense of belonging predicts mental health outcomes after accounting for other relevant factors. To address that question, we estimate two linear regression models: one for PHQ-9 depression scores and one for GAD-7 anxiety scores. In both cases, belonging is the predictor of interest, while academic year and financial stress are included as controls.

```python
if 'analysis_df' in locals():
    from math import erf, sqrt
    import numpy as np
    import pandas as pd

    def normal_sf(z):
        return 0.5 * (1 - erf(z / sqrt(2)))

    def fit_ols(data, y_col, x_cols):
        tmp = data[[y_col] + x_cols].dropna().copy()

        X = pd.get_dummies(tmp[x_cols], drop_first=True)
        X.insert(0, 'Intercept', 1.0)

        y = tmp[y_col].astype(float).values
        X_mat = X.astype(float).values

        beta = np.linalg.inv(X_mat.T @ X_mat) @ (X_mat.T @ y)
        fitted = X_mat @ beta
        resid = y - fitted

        n = len(y)
        p = X_mat.shape[1]
        dof = n - p

        sse = float(resid.T @ resid)
        mse = sse / dof
        cov_beta = mse * np.linalg.inv(X_mat.T @ X_mat)
        se = np.sqrt(np.diag(cov_beta))
        t_vals = beta / se
        p_vals = [2 * normal_sf(abs(t)) for t in t_vals]
        r2 = 1 - (sse / float(((y - y.mean()) ** 2).sum()))

        results = pd.DataFrame({
            'term': X.columns,
            'coefficient': beta,
            'std_error': se,
            't_value': t_vals,
            'approx_p_value': p_vals
        })

        return results, r2, n

    predictors = [flourish_col, fin_col, school_col]

    phq_results, phq_r2, phq_n = fit_ols(analysis_df, phq_col, predictors)
    gad_results, gad_r2, gad_n = fit_ols(analysis_df, gad_col, predictors)

    print('PHQ-9 model results:')
    display(phq_results)
    print(f'PHQ-9 model n = {phq_n}, R^2 = {phq_r2:.3f}')

    print('\nGAD-7 model results:')
    display(gad_results)
    print(f'GAD-7 model n = {gad_n}, R^2 = {gad_r2:.3f}')
```

```python
if 'analysis_df' in locals():
    clean_reg = analysis_df[[phq_col, gad_col, flourish_col, fin_col]].dropna().copy()

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    axes[0].scatter(clean_reg[flourish_col], clean_reg[phq_col], alpha=0.35)
    m1, b1 = np.polyfit(clean_reg[flourish_col], clean_reg[phq_col], 1)
    x1 = np.linspace(clean_reg[flourish_col].min(), clean_reg[flourish_col].max(), 100)
    axes[0].plot(x1, m1 * x1 + b1, color='red')
    axes[0].set_title('Flourishing Score vs PHQ-9')
    axes[0].set_xlabel('Flourishing score')
    axes[0].set_ylabel('PHQ-9 score')

    axes[1].scatter(clean_reg[flourish_col], clean_reg[gad_col], alpha=0.35)
    m2, b2 = np.polyfit(clean_reg[flourish_col], clean_reg[gad_col], 1)
    x2 = np.linspace(clean_reg[flourish_col].min(), clean_reg[flourish_col].max(), 100)
    axes[1].plot(x2, m2 * x2 + b2, color='red')
    axes[1].set_title('Flourishing Score vs GAD-7')
    axes[1].set_xlabel('Flourishing score')
    axes[1].set_ylabel('GAD-7 score')

    plt.tight_layout()
    plt.show()
else:
    print("analysis_df was not created.")
```

The overarching takeaway is that mental well-being is not just an isolated clinical issue; it is deeply intertwined with how integrated a student feels within their campus environment.

However, interpreting these results comes with some nuance, particularly when looking at why UC campuses might report slightly higher distress than the national baseline. It is highly possible that this difference isn't entirely due to a harsher academic environment. Campuses with aggressive mental health awareness campaigns often equip students with the vocabulary to identify their struggles and actively work to reduce the stigma of reporting them. As a result, higher average distress scores might partially reflect a greater willingness to self-report rather than an inherently unhealthier student body.

There are a few major limitations to keep in mind, mainly revolving around the survey methodology. Because participation is voluntary, self-selection bias is a real threat; students who are actively struggling with stress are likely more motivated to complete a mental health survey, potentially skewing the campus averages upward. Additionally, relying on cross-sectional data means we are only looking at a single snapshot in time, which makes it impossible to prove causality. We cannot say for sure whether a lack of belonging causes depression, or if experiencing depression causes a student to isolate and feel disconnected. If this investigation were to continue, the most valuable next step would be incorporating longitudinal data to track the same cohort of students from their freshman to senior year. It would also be worthwhile to stratify the data by specific majors to see if the rigorous demands of certain academic tracks are driving the slight variations we observed.
