## Abstract

This project investigates whether UC San Diego students differ in levels of depression and anxiety compared to other UC campuses and national benchmarks, and whether a student's sense of belonging predicts mental health outcomes. Using large-scale survey datasets including the Healthy Minds Study (HMS), ACHA–NCHA, and UCUES, we analyzed standardized measures such as PHQ-9, GAD-7, and belonging scores.

We first ensured data quality and comparability across datasets, then conducted descriptive and inferential analyses including prevalence estimation and regression modeling. Our models examined the relationship between belonging and mental health outcomes while controlling for academic year and financial stress.

The results show that depression and anxiety levels at UCSD are generally comparable to peer institutions, with slight variation depending on the benchmark used. More importantly, sense of belonging emerges as a strong and consistent predictor of both depression and anxiety, even after accounting for key confounders. Students with lower belonging scores tend to report significantly higher psychological distress.

Overall, these findings suggest that student mental health is not solely an individual clinical issue but is closely tied to campus environment and social integration. Improving students' sense of belonging may be an effective pathway for institutions to support mental well-being at scale.

## Research Question

Is there a statistically significant difference in the prevalence of clinical depression (PHQ-9 scores) and anxiety (GAD-7 scores) among UC San Diego undergraduates compared to the UC System-wide average and National US University benchmarks (as defined by the Healthy Minds Study 2024-2025)? Furthermore, to what extent does 'Sense of Belonging' (measured via the 6-point UCUES Likert scale) serve as a predictor for these mental health outcomes when controlling for academic year and financial stress.

### Dataset 1: Healthy Minds Study (HMS)

1. This data set is from the Healthy Minds Study (HMS), which is a large-scale survey adminstarted across US colleges and universities to assess the mental health of students. The data set includes student survey responses, including standardized mental health screening instruments, demographic characteristics, and data on use of counseling services. 2. Important metrics
        The primary mental health outcomes in this dataset include:

        Depression symptoms (PHQ-9 score)
        Unit: Score from 0–27
        Interpretation:
        0–4: Non - Minimal depression
        5–9: Mild depression
        10–14: Moderate depression
        15–19: Moderately severe

        Unit: Score from 0–21

        0–4: Non-Minimal anxiety

        Often measured on a scale of 1–5 or 0–10.

        Counseling service use
        Binary (0 = No, 1 = Yes) or categorical (Never, Past year, Past month).
3. Major Concerns:
    Self-report bias
    Students tend to underreport or overreport their symptoms based on stigma or personal awareness.
    Self-selection bias
    Voluntary participation. Students who choose to participate may systematically differ from those who do not participate (e.g., more distressed students tend to participate).
    Non-response bias
    Some groups (e.g., international students, students with heavy workload) tend to have lower response rates.
    Timing effects
    Surveys taken around the time of midterms or finals may artificially inflate levels of stress.

We examine the individual-level survey dataset containing student mental health indicators. A tidy data check confirms that the dataset follows a standard structure with one row per respondent. #### Clinical Prevalence and Group Mean Estimates for Depression and Anxiety

We also calculate the prevalence of elevated depression and anxiety symptoms using validated thresholds in the HMS dataset (PHQ-9 and GAD-7 cutoff >= 10).


```python
if 'df_hms' in locals():
    def find_col(columns, candidates):
        lower_map = {c.lower(): c for c in columns}
        for cand in candidates:
            if cand.lower() in lower_map:
                return lower_map[cand.lower()]
        return None

    phq_col = find_col(df_hms.columns, ['PHQ9_total', 'phq9_total', 'phq9', 'PHQ_9_total', 'PHQ9'])
    gad_col = find_col(df_hms.columns, ['GAD7_total', 'gad7_total', 'gad7', 'GAD_7_total', 'GAD7'])
    belong_col = find_col(df_hms.columns, ['belonging_score', 'sense_of_belonging', 'belonging', 'belonging_item'])
    school_col = find_col(df_hms.columns, ['school', 'institution', 'campus', 'school_name', 'institution_name'])
    
    if phq_col and gad_col:
        df_hms[phq_col] = pd.to_numeric(df_hms[phq_col], errors='coerce')
        df_hms[gad_col] = pd.to_numeric(df_hms[gad_col], errors='coerce')
        
        phq_clinical = df_hms[phq_col] >= 10
        gad_clinical = df_hms[gad_col] >= 10
        
        overall_prevalence = pd.DataFrame({
            'metric': ['Mean PHQ-9', 'Mean GAD-7', 'PHQ-9 >= 10', 'GAD-7 >= 10'],
            'value': [
                df_hms[phq_col].mean(),
                df_hms[gad_col].mean(),
                phq_clinical.mean(),
                gad_clinical.mean()
            ]
        })
        display(overall_prevalence)
        
        if school_col:
            school_summary = df_hms.groupby(school_col).agg(
                n=(phq_col, 'size'),
                mean_phq=(phq_col, 'mean'),
                mean_gad=(gad_col, 'mean')
            )
            school_summary['phq_ge_10'] = df_hms.groupby(school_col)[phq_col].apply(lambda s: (s >= 10).mean())
            school_summary['gad_ge_10'] = df_hms.groupby(school_col)[gad_col].apply(lambda s: (s >= 10).mean())
            display(school_summary.sort_values('n', ascending=False).head(10))
        
        fig, axes = plt.subplots(1, 2, figsize=(10, 4))
        overall_rates = pd.Series({'PHQ-9 >= 10': phq_clinical.mean(), 'GAD-7 >= 10': gad_clinical.mean()})
        overall_rates.plot(kind='bar', ax=axes[0])
        axes[0].set_title('Clinical-Range Prevalence')
        axes[0].set_ylabel('Proportion of students')
        
        axes[1].boxplot([df_hms[phq_col].dropna(), df_hms[gad_col].dropna()], labels=['PHQ-9', 'GAD-7'])
        axes[1].set_title('Symptom Score Spread')
        axes[1].set_ylabel('Score')
        plt.tight_layout()
        plt.show()
```

## Conclusion and Discussion

Our analysis aimed to determine how UC San Diego students' mental health compares to peer institutions and whether a student's sense of belonging reliably predicts psychological distress. The data revealed that depression and anxiety levels at UCSD are relatively consistent with the rest of the UC system, though they trend slightly higher than the national average. More importantly, the regression model highlighted a clear pattern: a lack of belonging on campus is a strong predictor of clinical distress. Even after factoring in heavy confounders like financial strain and class standing, students who felt disconnected were significantly more likely to report higher PHQ-9 and GAD-7 scores. The overarching takeaway is that mental well-being is not just an isolated clinical issue; it is deeply intertwined with how integrated a student feels within their campus environment.

These findings build directly on the foundation established by the Healthy Minds Study and ACHA-NCHA, reinforcing the idea that institutional culture plays a massive role in overall student health. However, interpreting these results comes with some nuance, particularly when looking at why UC campuses might report slightly higher distress than the national baseline. It is highly possible that this difference isn't entirely due to a harsher academic environment. Campuses with aggressive mental health awareness campaigns often equip students with the vocabulary to identify their struggles and actively work to reduce the stigma of reporting them. As a result, higher average distress scores might partially reflect a greater willingness to self-report rather than an inherently unhealthier student body.

There are a few major limitations to keep in mind, mainly revolving around the survey methodology. Because participation is voluntary, self-selection bias is a real threat; students who are actively struggling with stress are likely more motivated to complete a mental health survey, potentially skewing the campus averages upward. Additionally, relying on cross-sectional data means we are only looking at a single snapshot in time, which makes it impossible to prove causality. We cannot say for sure whether a lack of belonging causes depression, or if experiencing depression causes a student to isolate and feel disconnected. If this investigation were to continue, the most valuable next step would be incorporating longitudinal data to track the same cohort of students from their freshman to senior year. It would also be worthwhile to stratify the data by specific majors to see if the rigorous demands of certain academic tracks are driving the slight variations we observed.
