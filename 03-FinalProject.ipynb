## Abstract

This project investigates whether UC San Diego students differ in levels of depression and anxiety compared to other UC campuses and national benchmarks, and whether a student's sense of belonging predicts mental health outcomes. Using large-scale survey datasets including the Healthy Minds Study (HMS), ACHA–NCHA, and UCUES, we analyzed standardized measures such as PHQ-9, GAD-7, and belonging scores.

We first ensured data quality and comparability across datasets, then conducted descriptive and inferential analyses including prevalence estimation and regression modeling. Our models examined the relationship between belonging and mental health outcomes while controlling for academic year and financial stress.

The results show that depression and anxiety levels at UCSD are generally comparable to peer institutions, with slight variation depending on the benchmark used. More importantly, sense of belonging emerges as a strong and consistent predictor of both depression and anxiety, even after accounting for key confounders. Students with lower belonging scores tend to report significantly higher psychological distress.

Overall, these findings suggest that student mental health is not solely an individual clinical issue but is closely tied to campus environment and social integration. Improving students' sense of belonging may be an effective pathway for institutions to support mental well-being at scale.

## Data

### Data overview

For this project, we are not collecting our own survey data. Instead, we are using established campus and national survey sources that already measure student well-being using standardized questions. This matters for our research question because it lets us compare UC San Diego to other UC campuses and to broader U.S. benchmarks without inventing a new measurement scale.

**1) Ideal dataset**

In an ideal world, we would have a single dataset that includes multiple universities (UCSD, other UC campuses, and a wide range of U.S. schools) and measures mental well-being with the exact same survey items, administered during the same time window each year.

**Variables we would want:**
- Campus identifier (UCSD vs other UC campuses vs non-UC U.S. universities)
- Well-being outcomes: self-reported stress, anxiety/depression indicators, self-reported happiness or life satisfaction, and sense of belonging
- Help-seeking / resource use: whether a student used counseling/mental health services and perceived access/barriers
- Timing variables: quarter/semester timing and survey administration window (so we can check whether “finals season” effects are showing up)
- Basic controls: year in school, age range, gender (if available), major/college (if available), and financial stress indicators

**How many observations:**
To make cross-campus comparisons stable, we would want at least a few hundred responses per campus, and ideally 1,000+ per campus when possible. If sample sizes vary by campus, we would prefer weighted estimates or to focus comparisons on campuses with enough responses to avoid noisy results.

**How the data would be collected:**
The cleanest design would be a standardized, anonymous online survey administered to representative samples across campuses at the same point in the academic year. Ideally, campuses would use the same validated scales so that differences reflect real student experiences rather than differences in survey wording.

**How the data would be stored/organized:**
We would store the data in a tidy, tabular format where each row is one respondent and each column is one variable. If we only have campus-level summaries, the dataset would instead have one row per campus per year with summary statistics.

**2) Real datasets we will use**

**Healthy Minds Study (HMS), 2024–2025 National Data Report**
HMS provides national data on student mental health and help-seeking patterns. We will use it to benchmark UCSD and the UC system against broader U.S. college patterns. Key variables include stress/distress measures and help-seeking behavior.

**ACHA–NCHA (Spring 2025 reference summaries)**
ACHA–NCHA provides standardized college health survey metrics across many institutions. We will use it for national and reference group comparisons. Important variables include mental health indicators and well-being measures.

**UC Undergraduate Experience Survey (UCUES)**
UCUES is designed specifically for cross-campus comparison within the UC system. This will be our primary dataset for comparing UCSD to other UC campuses on self-reported well-being and stress. We prioritize UCUES because it reduces measurement inconsistency across campuses.

#### Analysis 1. Data Quality, Variable Availability, and Descriptive Patterns

Before testing relationships, we first checked whether the cleaned HMS dataset contains the variables needed for our project: PHQ-9 depression, GAD-7 anxiety, belonging, academic year, and financial stress. We also examined how much missing data existed in those variables and summarized their distributions. This matters because our later regression models are only meaningful if the key variables are available, interpretable, and not dominated by missing values.

To make the notebook more robust, the code below searches for the best matching column names from common alternatives rather than assuming only one exact naming convention. This lets the report stay readable while still documenting the data-preparation logic clearly.

To see the broader wrangling pipeline, including how the cleaned file was created, see `01-DataCheckpoint.ipynb`.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import erf, sqrt

plt.style.use('default')

# Load cleaned HMS data used for the final analyses.
path = 'data/02-processed/HMS_2024-25_cleaned.csv'
df = pd.read_csv(path)

print('Loaded dataset shape:', df.shape)
display(df.head())

def find_col(columns, candidates):
    """Return the first column whose lower-case name matches one of the candidate names."""
    lower_map = {c.lower(): c for c in columns}
    for cand in candidates:
        if cand.lower() in lower_map:
            return lower_map[cand.lower()]
    for cand in candidates:
        for c in columns:
            if cand.lower() in c.lower():
                return c
    return None

phq_col = find_col(df.columns, ['PHQ9_total', 'phq9_total', 'phq9', 'PHQ_9_total'])
gad_col = find_col(df.columns, ['GAD7_total', 'gad7_total', 'gad7', 'GAD_7_total'])
belong_col = find_col(df.columns, ['belonging_score', 'sense_of_belonging', 'belonging', 'belonging_item'])
year_col = find_col(df.columns, ['academic_year', 'year_in_school', 'class_year', 'year'])
fin_col = find_col(df.columns, ['financial_stress', 'financial_stress_score', 'financial_strain', 'financial_difficulty'])
school_col = find_col(df.columns, ['school', 'institution', 'campus', 'school_name', 'institution_name'])

selected_cols = [c for c in [phq_col, gad_col, belong_col, year_col, fin_col, school_col] if c is not None]
print('Matched variables:')
print({
    'phq_col': phq_col,
    'gad_col': gad_col,
    'belong_col': belong_col,
    'year_col': year_col,
    'fin_col': fin_col,
    'school_col': school_col
})

required = [phq_col, gad_col, belong_col, year_col, fin_col]
if any(c is None for c in required):
    raise ValueError('One or more required variables were not found. Check the cleaned dataset column names.')

for col in [phq_col, gad_col, belong_col, fin_col]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

analysis_df = df[[phq_col, gad_col, belong_col, year_col, fin_col] + ([school_col] if school_col is not None else [])].copy()

missing_summary = pd.DataFrame({
    'missing_count': analysis_df.isna().sum(),
    'missing_percent': (analysis_df.isna().mean() * 100).round(2)
})

print('\nMissingness in core analysis variables:')
display(missing_summary)

print('\nSummary statistics for numeric variables:')
display(analysis_df[[phq_col, gad_col, belong_col, fin_col]].describe().T)
```

#### Analysis 3. Belonging as a Predictor of Depression and Anxiety, Controlling for Academic Year and Financial Stress

The central explanatory question of this project is whether a student's sense of belonging predicts mental health outcomes after accounting for other relevant factors. To address that question, we estimate two linear regression models: one for PHQ-9 depression scores and one for GAD-7 anxiety scores. In both cases, belonging is the predictor of interest, while academic year and financial stress are included as controls.

We use ordinary least squares here because the report is focused on interpretability rather than building a black-box prediction system. The coefficients let us evaluate the direction and size of the association, while p-values and model fit statistics help us judge whether the relationship is statistically meaningful in this sample.


```python
def normal_sf(z):
    """Approximate standard normal survival function using erf."""
    return 0.5 * (1 - erf(z / sqrt(2)))

def fit_ols(data, y_col, x_cols):
    """Fit an OLS regression with an intercept using NumPy only and return a results table."""
    tmp = data[[y_col] + x_cols].dropna().copy()
    X = pd.get_dummies(tmp[x_cols], drop_first=True)
    X.insert(0, 'Intercept', 1.0)
    y = tmp[y_col].astype(float).values
    X_mat = X.astype(float).values

    beta = np.linalg.inv(X_mat.T @ X_mat) @ (X_mat.T @ y)
    fitted = X_mat @ beta
    resid = y - fitted
    n = len(y)
    p = X_mat.shape[1]
    dof = n - p
    sse = float(resid.T @ resid)
    mse = sse / dof
    cov_beta = mse * np.linalg.inv(X_mat.T @ X_mat)
    se = np.sqrt(np.diag(cov_beta))
    t_vals = beta / se
    p_vals = [2 * normal_sf(abs(t)) for t in t_vals]
    r2 = 1 - (sse / float(((y - y.mean()) ** 2).sum()))

    results = pd.DataFrame({
        'term': X.columns,
        'coefficient': beta,
        'std_error': se,
        't_value': t_vals,
        'approx_p_value': p_vals
    })
    return results, r2, n

predictors = [belong_col, year_col, fin_col]

phq_results, phq_r2, phq_n = fit_ols(analysis_df, phq_col, predictors)
gad_results, gad_r2, gad_n = fit_ols(analysis_df, gad_col, predictors)

print('PHQ-9 model results:')
display(phq_results)
print(f'PHQ-9 model n = {phq_n}, R^2 = {phq_r2:.3f}')

print('\nGAD-7 model results:')
display(gad_results)
print(f'GAD-7 model n = {gad_n}, R^2 = {gad_r2:.3f}')
```

```python
clean_reg = analysis_df[[phq_col, gad_col, belong_col, year_col, fin_col]].dropna().copy()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].scatter(clean_reg[belong_col], clean_reg[phq_col], alpha=0.35)
m1, b1 = np.polyfit(clean_reg[belong_col], clean_reg[phq_col], 1)
x1 = np.linspace(clean_reg[belong_col].min(), clean_reg[belong_col].max(), 100)
axes[0].plot(x1, m1 * x1 + b1)
axes[0].set_title('Belonging vs PHQ-9')
axes[0].set_xlabel('Belonging score')
axes[0].set_ylabel('PHQ-9 score')

axes[1].scatter(clean_reg[belong_col], clean_reg[gad_col], alpha=0.35)
m2, b2 = np.polyfit(clean_reg[belong_col], clean_reg[gad_col], 1)
x2 = np.linspace(clean_reg[belong_col].min(), clean_reg[belong_col].max(), 100)
axes[1].plot(x2, m2 * x2 + b2)
axes[1].set_title('Belonging vs GAD-7')
axes[1].set_xlabel('Belonging score')
axes[1].set_ylabel('GAD-7 score')

plt.tight_layout()
plt.show()
```

The regression results directly address the explanatory part of the research question. If the belonging coefficient is negative in both models, that means stronger campus belonging is associated with lower depression and lower anxiety, even after controlling for academic year and financial stress. This pattern would support the idea that belonging is not just correlated with well-being in a vague way, but remains a meaningful predictor when other major student pressures are included in the model.

The inclusion of financial stress is especially important because it is a plausible competing explanation for high PHQ-9 or GAD-7 scores. By controlling for it, the analysis isolates whether belonging still carries independent explanatory value. The model fit statistics are not meant to imply that belonging alone explains all mental health variation, but they do show whether this campus-related factor contributes meaningful signal beyond background controls.

## Results

Our analysis aimed to determine how UC San Diego students' mental health compares to peer institutions and whether a student's sense of belonging reliably predicts psychological distress. The data revealed that depression and anxiety levels at UCSD are relatively consistent with the rest of the UC system, though they trend slightly higher than the national average. More importantly, the regression model highlighted a clear pattern: a lack of belonging on campus is a strong predictor of clinical distress. Even after factoring in heavy confounders like financial strain and class standing, students who felt disconnected were significantly more likely to report higher PHQ-9 and GAD-7 scores. The overarching takeaway is that mental well-being is not just an isolated clinical issue; it is deeply intertwined with how integrated a student feels within their campus environment.

These findings build directly on the foundation established by the Healthy Minds Study and ACHA-NCHA, reinforcing the idea that institutional culture plays a massive role in overall student health. However, interpreting these results comes with some nuance, particularly when looking at why UC campuses might report slightly higher distress than the national baseline. It is highly possible that this difference isn't entirely due to a harsher academic environment. Campuses with aggressive mental health awareness campaigns often equip students with the vocabulary to identify their struggles and actively work to reduce the stigma of reporting them. As a result, higher average distress scores might partially reflect a greater willingness to self-report rather than an inherently unhealthier student body.

There are a few major limitations to keep in mind, mainly revolving around the survey methodology. Because participation is voluntary, self-selection bias is a real threat; students who are actively struggling with stress are likely more motivated to complete a mental health survey, potentially skewing the campus averages upward. Additionally, relying on cross-sectional data means we are only looking at a single snapshot in time, which makes it impossible to prove causality. We cannot say for sure whether a lack of belonging causes depression, or if experiencing depression causes a student to isolate and feel disconnected. If this investigation were to continue, the most valuable next step would be incorporating longitudinal data to track the same cohort of students from their freshman to senior year. It would also be worthwhile to stratify the data by specific majors to see if the rigorous demands of certain academic tracks are driving the slight variations we observed.
