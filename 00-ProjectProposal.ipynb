Research question: How does the mental wellbeing and overall happiness of a student at UCSD compare to those in other UC’s? Other United States universities? Compared to universities around the world?

## Background and Prior Work


Students' mental health has become a central concern in the field of higher education, because psychological distress affects not only students' quality of life, but also their academic performance and their behaviors to seek help. To monitor indications such as depression symptoms, anxiety, stress, sense of belonging, and usage of counseling  service, campuses use standardized survey tools. However, different campuses measure mental health and well-being in different ways. As a result, we must carefully design metric selection and consider survey design variation, including sampling strategies, response rates and timing within a academic year, these are necessary for the studies.

This project will focus on asking how UCSD students' mental well-being and overall happiness compare to other UC campuses, other US universities, and univeristy internationlly. It can be framed as descriptive analysis(UCSD vs other UC campuses) and explanatory modeling(what student characteristics or campus conditions predict differences, after adjusting for confounders like student demographics, financial stress, or academic workload).
There are prior researches and work show that student mental health issues are common and institutional variables can affect the prevalence of symtoms and the likelyhood of seeking help.

Healthy Minds Study(HMS)<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

One of the most widely used systems for comparing mental health status in US colleges and universities is the Healthy Minds Study (HMS).The study surveyed students at participating institutions and reported prevalence rates of mental health indicators, help-seeking behaviors, and demand for services.Results from HMS indicate that mental health challenges are common among US college students. More importantly, the study shows a consistent gap between the number of students experiencing mental health difficulties and the number who actually seek treatment. This suggests that differences between campuses may reflect variations in access to care, stigma, or awareness, rather than differences in mental health alone.

ACHA-NCHA<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

This is one of the largest and most comprehensive surveys of college student health in the United States. Conducted over multiple years across numerous institutions, the reference group summaries released by ACHA enable colleges to compare their student data against national benchmarks. These summaries encompass assessment metrics related to mental health, well-being, sense of belonging, and health behaviors.

Research results indicate that well-being is influenced both by mental health status and by broader campus experiences, such as a sense of belonging and perceptions of institutional support. This supports a perspective that views well-being as multidimensional rather than solely focused on mental health symptoms.


UC Undergraduate Experience Survey (UCUES) data<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

For comparison between UC campuses, one of the most practical tool is the UCUES, the UC Undergraduate Experience Survey, it publishes campus-comparable tables for undergraduates, which can support UCSD vs UCLA/UCB/etc. comparisons using consistent question wording and timing.

Further more, the UC systems provide Student Health & Wellbeing dashboard that presents UC health/wellbeing data collected through the NCHA instrument, along with documented data download instructions for the dashboard outputs.

This tool prevent the biggest methodological obstacle in cross-campuses research: measurement inconsistency.
